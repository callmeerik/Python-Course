# üìñ Introduction to Seaborn

Seaborn is a Python data visualization library based on **matplotlib**. It provides a high-level interface for creating **informative and attractive statistical graphics**.

### ‚ú® Key Features:
- Built on top of Matplotlib üé®
- Beautiful default styles üíÖ
- Easy to use with pandas DataFrames üêº
- Special focus on statistical plots üìä
- Supports categorical and numerical plots
- Provides complex visualizations with very little code

## üìä Cheat Sheet Table

| Graph Name | Variable Type | Use Case | Example Statement | Code Formula |
|------------|---------------|----------|------------------|--------------|
| Bar Plot | Categorical vs Numerical | Compare categories by a numeric measure | Sales per pizza type | `sns.barplot(x='pizza_type', y='sales', data=df)` |
| Count Plot | Categorical | Show frequency of categories | Count of orders by day | `sns.countplot(x='day_of_week', data=df)` |
| Histogram | Numerical | Distribution of a single numeric variable | Distribution of delivery times | `sns.histplot(df['delivery_time'])` |
| Boxplot | Categorical vs Numerical | Show distribution & outliers | Pizza price distribution by type | `sns.boxplot(x='pizza_type', y='price', data=df)` |
| Violin Plot | Categorical vs Numerical | Show distribution + KDE | Delivery time distribution per day | `sns.violinplot(x='day_of_week', y='delivery_time', data=df)` |
| Scatter Plot | Numerical vs Numerical | Show relationship between 2 numeric vars | Price vs Sales | `sns.scatterplot(x='price', y='sales', data=df)` |
| Line Plot | Numerical vs Numerical | Trend over time/continuous variable | Sales by day | `sns.lineplot(x='day_of_week', y='sales', data=df)` |
| Heatmap | Both (matrix) | Show correlation or density in matrix form | Correlation between variables | `sns.heatmap(df.corr(), annot=True)` |
| Pairplot | Multiple numerical | Visualize pairwise relationships | Relations among price, sales, rating | `sns.pairplot(df[['price','sales','rating']])` |
| FacetGrid | Categorical + Any | Compare plots across categories | Histograms of sales by pizza type | `sns.FacetGrid(df, col='pizza_type').map(sns.histplot, 'sales')` |
| Subplots (plt.subplots) | Any | Multiple plots side by side | Sales vs Price and Histogram of Delivery Times | `fig, ax = plt.subplots(1,2)` ...

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# üçï Pizza Dataset
np.random.seed(42)  # reproducibilidad

pizza_types = ['Margherita','Pepperoni','Hawaiian','Veggie','BBQ Chicken']
days = ['Mon','Tue','Wed','Thu','Fri']

data = {
    'pizza_type': np.random.choice(pizza_types, 20),
    'price': np.random.randint(7, 12, 20),  # precios entre 7 y 11
    'sales': np.random.randint(30, 80, 20), # ventas entre 30 y 80
    'day_of_week': np.random.choice(days, 20),
    'delivery_time': np.random.randint(25, 45, 20),  # tiempo de entrega
    'rating': np.random.randint(3, 6, 20)  # rating 3 a 5
}
df = pd.DataFrame(data)
df.head()
```

## üìä Bar Plot

```python
sns.barplot(x='pizza_type', y='sales', data=df)
plt.title('Sales per Pizza Type')
plt.xlabel('Pizza Type')
plt.ylabel('Sales')
plt.show()
```

## üìä Count Plot

```python
sns.countplot(x='day_of_week', data=df)
plt.title('Number of Orders per Day')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()
```

## üìä Histogram

```python
sns.histplot(df['price'], kde=True, bins=10)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## üìä Boxplot

```python
sns.boxplot(x='pizza_type', y='price', data=df)
plt.title('Price Distribution by Pizza Type')
plt.xlabel('Pizza Type')
plt.ylabel('Price')
plt.show()
```

## üìä Violin Plot

```python
sns.violinplot(x='day_of_week', y='delivery_time', data=df)
plt.title('Delivery Time by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Delivery Time (minutes)')
plt.show()
```

## üìä Scatter Plot

```python
sns.scatterplot(x='price', y='sales', data=df)
plt.title('Price vs Sales')
plt.xlabel('Price')
plt.ylabel('Sales')
plt.show()
```

## üìä Line Plot

```python
sns.lineplot(x='day_of_week', y='sales', data=df)
plt.title('Sales by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Sales')
plt.show()
```

## üìä Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## üìä Pairplot

```python
sns.pairplot(df[['price','sales','rating']])
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

## üìä FacetGrid

```python
g = sns.FacetGrid(df, col='pizza_type', col_wrap=3)
g.map(sns.histplot, 'sales')
plt.show()
```

## üìä Subplots with plt.subplots()

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Scatter on first subplot
sns.scatterplot(x='price', y='sales', data=df, ax=axes[0])
axes[0].set_title('Price vs Sales')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Sales')

# Histogram on second subplot
sns.histplot(df['delivery_time'], kde=True, ax=axes[1])
axes[1].set_title('Delivery Time Distribution')
axes[1].set_xlabel('Minutes')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
