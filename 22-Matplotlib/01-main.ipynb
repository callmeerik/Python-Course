# ðŸ“Š Introduction to Matplotlib
Matplotlib is a powerful Python library for creating static, animated, and interactive visualizations. It is widely used in data science and analytics for producing publication-quality plots.

**Main characteristics:**
- Provides fine-grained control over plots.
- Supports a wide variety of plots: bar, line, scatter, histogram, pie, etc.
- Highly customizable: colors, labels, titles, annotations.
- Integrates with NumPy and Pandas seamlessly.

## ðŸ“‘ Cheat Sheet
| Graph Name   | Variable Type      | Use Case / Interpretation | Example Statement | Code Formula |
|--------------|--------------------|----------------------------|-------------------|--------------|
| Bar Plot     | Categorical vs Numerical | Compare categories | Compare pizza sales by type | plt.bar(categories, values) |
| Histogram    | Numerical          | Distribution of values | Distribution of delivery times | plt.hist(values, bins=...) |
| Scatter Plot | Numerical vs Numerical | Relationship between variables | Relation between pizza price and sales | plt.scatter(x, y) |
| Pie Chart    | Categorical        | Proportion of categories | Percentage of sales by pizza type | plt.pie(values, labels=...) |
| Line Plot    | Numerical or Time  | Trend over time | Sales over days of the week | plt.plot(x, y) |
| Boxplot      | Numerical          | Spread and outliers | Price distribution of pizzas | plt.boxplot(values) |

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset for a pizzeria
data = {
    'pizza_type': ['Margherita','Pepperoni','Hawaiian','Veggie','BBQ Chicken','Four Cheese'],
    'price': [8, 10, 9, 8, 11, 10],
    'sales': [120, 200, 80, 95, 150, 110],
    'day_of_week': ['Mon','Tue','Wed','Thu','Fri','Sat'],
    'delivery_time': [25, 30, 28, 35, 40, 20],
    'rating': [4.2, 4.5, 3.8, 4.1, 4.3, 4.0]
}
df = pd.DataFrame(data)
df
```

## ðŸ“Š Bar Plot

```python
plt.bar(df['pizza_type'], df['sales'])
plt.title('Pizza Sales by Type')
plt.xlabel('Pizza Type')
plt.ylabel('Units Sold')
plt.show()
```

## ðŸ“Š Histogram

```python
plt.hist(df['delivery_time'], bins=5, color='orange')
plt.title('Distribution of Delivery Times')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()
```

## ðŸ“Š Scatter Plot

```python
plt.scatter(df['price'], df['sales'], color='green')
plt.title('Price vs Sales')
plt.xlabel('Price ($)')
plt.ylabel('Units Sold')
plt.show()
```

## ðŸ“Š Pie Chart

```python
plt.pie(df['sales'], labels=df['pizza_type'], autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Sales by Pizza Type')
plt.show()
```

## ðŸ“Š Line Plot

```python
plt.plot(df['day_of_week'], df['sales'], marker='o')
plt.title('Sales by Day of the Week')
plt.xlabel('Day of Week')
plt.ylabel('Units Sold')
plt.show()
```

## ðŸ“Š Boxplot

```python
plt.boxplot(df['price'])
plt.title('Distribution of Pizza Prices')
plt.ylabel('Price ($)')
plt.show()
```
